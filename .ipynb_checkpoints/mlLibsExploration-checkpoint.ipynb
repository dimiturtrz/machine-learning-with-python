```python
data = original.copy()
```

```python
data.dtypes
```

```python
X = data[["satisfaction_level", "last_evaluation", "number_project", "average_montly_hours",
          "time_spend_company", "Work_accident", "promotion_last_5years"]]
y = data["left"]
# най-елементарното нещо - взимам само колоните с числа (salary може да вкара някаква допълнителна точност, но засега го изпускаме)
```

```python
X.dtypes
```

```python
y.value_counts() # това, което видяхме и на графиката
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y) 
# stratify може би е голям момент тук, понеже dataset-а е небалансиран
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression().fit(X_train, y_train)
print("train score:", model.score(X_train, y_train))
print("test score: ", model.score(X_test, y_test))

# линейната регресия се справя горе-долу добре
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier().fit(X_train, y_train)
print("train score:", model.score(X_train, y_train))
print("test score: ", model.score(X_test, y_test))

# random forest-a се справя изненадващо добре, въпреки, че може да има някакъв оувърфит заради небалансираността
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression

randomForest = RandomForestClassifier()

scores = cross_val_score(randomForest, X, y, cv=10, scoring='accuracy')
print(scores)
print("Mean: {}".format(scores.mean()))
print("Std:  {}".format(scores.std()))

# някакъв грид-сърч за да намерим все пак най-добрия, изненадващо, baseline модел
```
